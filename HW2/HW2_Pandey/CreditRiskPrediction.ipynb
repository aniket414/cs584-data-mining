```python
from sklearn.preprocessing import LabelEncoder
le_train = LabelEncoder()
le_test = LabelEncoder()

for column in ["Race", "Gender"]:
    le_train.fit(dataset_train[column])
    dataset_train[column] = le_train.transform(dataset_train[column])
    
    le_test.fit(dataset_test[column])
    dataset_test[column] = le_test.transform(dataset_test[column])

dataset_train.head()
dataset_test.head()
```

```python
X = dataset_train.iloc[:, 1:-1].values
y = dataset_train.iloc[:, -1].values
Actual_X_test = dataset_test.iloc[:, 1:12].values
```

```python
print(X.shape)
print(y.shape)
print(Actual_X_test.shape)
```

```python
uniqueData, dataCounts = np.unique(y, return_counts = True)
print("Data imbalance before sampling: ", dict(zip(uniqueData, dataCounts)))

randomUnderSampler = RandomUnderSampler()
X, y = randomUnderSampler.fit_resample(X, y)

uniqueData, dataCounts = np.unique(y, return_counts = True)
print("Data imbalance after sampling: ", dict(zip(uniqueData, dataCounts)))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
Models = {
    "SVC": SVC(kernel = 'linear', random_state = 0),
    "DecisionTree": DecisionTreeClassifier(criterion = 'entropy', random_state = 0),
    "RandomForest": RandomForestClassifier(criterion = 'gini', max_depth = 10, n_estimators = 500),
    "GaussianNaiveBayes": GaussianNB(),
    "KNeighborsClassifier": KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
}
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_cv = sc.fit_transform(X_train)
X_test_cv = sc.transform(X_test)

from sklearn.model_selection import cross_validate
cross_validation_results = pd.DataFrame(columns = ['model', 'train_score', 'test_score'])

for key in Models.keys():
    cross_validation_result = cross_validate(Models[key], X_train_cv, y_train, 
                                             return_train_score = True, scoring = "f1", cv = 5, n_jobs = -1)
    result = {
        'model': key, 
        'train_score': cross_validation_result["train_score"].mean(), 
        'test_score': cross_validation_result["test_score"].mean(),
        'fit_time': cross_validation_result["fit_time"].mean(),
        'score_time': cross_validation_result["score_time"].mean(),
        }
    cross_validation_results = cross_validation_results.append(result, ignore_index = True)
cross_validation_results
```

```python
print(cross_validation_results)
```

```python
randomForest = Models["RandomForest"].fit(X_train, y_train)
y_pred_randomForest = randomForest.predict(X_test)
print('F1 Score:', f1_score(y_test, y_pred_randomForest))
print(confusion_matrix(y_test, y_pred_randomForest))
print('Accuracy:', accuracy_score(y_test, y_pred_randomForest))
```

```python
svm = Models["SVC"].fit(X_train_cv, y_train)
y_pred_svm = svm.predict(X_test_cv)
print('F1 Score:', f1_score(y_test, y_pred_svm))
print(confusion_matrix(y_test, y_pred_svm))
print('Accuracy:', accuracy_score(y_test, y_pred_svm))
```

```python
decisionTree = Models["DecisionTree"].fit(X_train, y_train)
y_pred_decisionTree = decisionTree.predict(X_test)
print('F1 Score:', f1_score(y_test, y_pred_decisionTree))
print(confusion_matrix(y_test, y_pred_decisionTree))
print('Accuracy:', accuracy_score(y_test, y_pred_decisionTree))
```

```python
gaussianNaiveBayes = Models["GaussianNaiveBayes"].fit(X_train, y_train)
y_pred_gaussianNaiveBayes = gaussianNaiveBayes.predict(X_test)
print('F1 Score:', f1_score(y_test, y_pred_gaussianNaiveBayes))
print(confusion_matrix(y_test, y_pred_gaussianNaiveBayes))
print('Accuracy:', accuracy_score(y_test, y_pred_gaussianNaiveBayes))
```

```python
kNeighborsClassifier = Models["KNeighborsClassifier"].fit(X_train, y_train)
y_pred_kNeighborsClassifier = kNeighborsClassifier.predict(X_test)
print('F1 Score:', f1_score(y_test, y_pred_kNeighborsClassifier))
print(confusion_matrix(y_test, y_pred_kNeighborsClassifier))
print('Accuracy:', accuracy_score(y_test, y_pred_kNeighborsClassifier))
```

```python
def featureImportance(classifier, isTree = False):
    feature_importance = pd.DataFrame()
    feature_importance["feature"] = dataset_train.keys()
    
    #Drop Id and Credit Column
    feature_importance.drop(feature_importance.tail(1).index, inplace = True)
    feature_importance.drop(feature_importance.head(1).index, inplace = True)
    
    if isTree == True:
        feature_importance["importance"] = classifier.feature_importances_
    else:
        feature_importance["importance"] = classifier.coef_[0]
    
        
    feature_importance = feature_importance.sort_values("importance", ascending = False)
    print(feature_importance, '\n\n')
```

```python
print('RandomForest\n')
featureImportance(randomForest, isTree = True)
print('\nSVM\n')
featureImportance(svm)
print('\nDecisionTree\n')
featureImportance(decisionTree, isTree = True)
```
